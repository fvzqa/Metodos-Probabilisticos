```r
nRows <- 50
nCols <- 100

myMatrix <- matrix(runif(nRows*nCols, 0, 100), nrow = nRows, ncol = nCols)
kernel <- matrix(c(-1,-1,0,-1,1,1,0,1,1), nrow = 3, ncol = 3)

# Make convolution
myOutput1 <- convolution2D(myMatrix, kernel)
myOutput2 <- convolution2D(myOutput1, kernel)

# Plot results
par(mfrow = c(2, 2))
image(myMatrix, zlim = c(0, 100))
image(myOutput1, zlim = c(0, 100))
image(myOutput2, zlim = c(0, 100))
```

```r
teo2 <- function (X,Y){
    a <- round(var(X+Y),digits = 14)
    b <- round(var(X) + var(Y) + 2*cov(X,Y), digits = 14)
    return(a==b)
```

```r
resultados <- c()
for (i in 1:100){
    X <- rnorm(5000)
    Y <- runif(5000)
    resultados[i] <- teo2(X,Y)
}
sum(resultados)
```

```r
resultados <- c()
for (i in 1:100){
    X <- rnorm(5000)
    Y <- rexp(5000,1)
    resultados[i] <- teo2(X,Y)
}
sum(resultados)
```

```r
resultados <- c()
for (i in 1:100){
    X <- rexp(5000,1)
    Y <- rpois(5000,2)
    resultados[i] <- teo2(X,Y)
}
sum(resultados)
```

```r
teo1 <- function (a,X,b,c,Y,d){
    m <- round(cov(a*X+b, c*Y+d), digits = 14)
    n <- round(a*c*cov(X,Y), digits=14)
    return(m==n)
}
```

```r
resultados <- c()
for (i in 1:100){
    a <- runif(1)
    b <- runif(1)
    c <- runif(1)
    d <- runif(1)
    X <- rnorm(5000)
    Y <- runif(5000)
    resultados[i] <- teo1(a,X,b,c,Y,d)
}
sum(resultados)
```
