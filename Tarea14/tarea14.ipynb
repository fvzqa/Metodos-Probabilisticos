### Caminata aleatoria 1D

```r
pasos <- 10000

actual <- 0

posiciones <- c(actual)

p <- .5
```

```r
for (i in 1:pasos){
    u <- runif(1)
    if (u <= p){paso <- 1}
    else{ paso <- -1}
    actual <- actual + paso
    posiciones <- c(posiciones, actual)
}
```

```r
plot(posiciones)
```

```r
posiciones_simetrica <- posiciones
```

```r
plot(posiciones_simetrica, type='o', col="aquamarine", xlab="Paso", ylab="Posicin")
abline(h=0, col="red")
```

```r
png(filename="ca_p5.png", height=2000, width=2000, res=400)
plot(posiciones_simetrica, type='o', col="aquamarine", xlab="Paso", ylab="Posicin")
abline(h=0, col="red")
dev.off()
```

### Caminata aleatoria no discreta

```r
pasos <- 1000

ultima_posicion <- c()
```

```r
for (i in 1:1000){
    actual <- 0
    for ( i in 1:pasos){
        dx <- rexp(1)
        actual <- actual + dx
    }

    ultima_posicion <- c(ultima_posicion, actual)
}
```

```r
hist(ultima_posicion, col="plum3", main="", xlab="ltima posicin", ylab="Frecuencia")
```

```r
png(filename="ultima_posicion.png", height=2000, width=2000, res=400)
hist(ultima_posicion, col="plum3", main="", xlab="ltima posicin", ylab="Frecuencia")
dev.off()
```

```r
hist(rnorm(n = 1000, mean = 1000, sd = sqrt(1000)), col="lightblue", main="", xlab="Nmero", ylab="Frecuencia")
```

```r
png(filename="hist_normal.png", height=2000, width=2000, res=400)
hist(rnorm(n = 1000, mean = 1000, sd = sqrt(1000)), col="lightblue", main="", xlab="Nmero", ylab="Frecuencia")
dev.off()
```
